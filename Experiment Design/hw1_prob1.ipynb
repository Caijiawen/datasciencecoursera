```python
import numpy as np
from sympy import *
b0, b1, x = symbols("b0, b1, x")
b = Matrix([[b0-16], [b1-7.5]])
b_t = Matrix([[b0-16,b1-7.5]])

g = Matrix([[1],[x]])
g_t = Matrix([[1,x]])

def makeInformationMatrix(array):
    return Matrix([[1,np.mean(array)],[np.mean(array),np.mean(array*array)]])
```

```python
design = [np.array([-1,0,1]) , 
          np.array([-1,-1,0,0,1,1]) , 
          np.array([-1,-1,-1,-1,-1,-1,1,1]) , 
          np.array([-1,-0.5,0,0.5,1]) , 
          np.array([-1,-1,-0.9,-0.85,-0.8,-0.75,1]) , 
          np.array([-1,1])]
```

```python
#置信椭球
for array in design:
    expression1 = b_t *makeInformationMatrix(array) * b 
    print expression1
```

```python
#标准差方差
for array in design:
    expression2 = g_t *(makeInformationMatrix(array).inv()) * g 
    print expression2
```
